```python
import pandas as pd
df=pd.read_csv('new_homeprices.csv')
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
# merging "df" and "dummies"
```

```python
final=merged.drop(['town','west windsor'],axis='columns')
final
# "drop" doesnt by default delete that particular column....it returns a new dataframe without that column.....
# here i dont want columns like "columns" and "west windsor"
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x=final.drop(['price'], axis='columns')
x
# here my x coordinate is everything except "price"
```

```python
y=final.price
y
```

```python
model.fit(x,y)
#  now in future you may have a lot of data.....and that could take hours to get trained.......so what you do is:=
# after training save it as model
```

```python
model.predict([[2800,0,1]])

# as my "x" coordinate has 3 fields...."area","monroes township" and "robinsville"
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(x,y)
# it is very important method which allows u to check the correctness of ur model.....
# it will create a table with predicted values and check it with given  table and return a score
```
